```python
def clean_price(price):
  return float("".join(char for char in price if char.isdigit() or char == ""))
```

```python
def clean_rating(rating):
  rating_map ={
      'One':1,
      "Two":2,
      "Three":3,
      "Four":4,
      "Five":5
  }
  return rating_map[rating]
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com/'
results = requests.get(url)
results.status_code

soup = BeautifulSoup(results.content, 'html.parser')
books = soup.find_all('article', class_='product_pod')

def book_extract(book):
    title = book.h3.a.attrs['title']
    price = book.find("p",class_='price_color').get_text()
    rating = book.find("p",class_='star-rating')['class'][1]
    stock = book.find("p",class_='instock availability').get_text().strip()
    return{
        'Title':title,
        "Price":clean_price(price),
        'Rating':clean_rating(rating),
        "Stock":stock
    }

df =[book_extract(book) for book in books]
```
